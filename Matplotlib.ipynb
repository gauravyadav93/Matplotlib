```python
import pandas as pd
import matplotlib.pyplot as plt
world = pd.read_csv('Python Course/Data for Python Course/2018-2019 Happiness.csv') #importing world hapiness dataset
airbnb = pd.read_csv('Python Course/Data for Python Course/airbnb.csv') #importing airbnb dataset
salaries = pd.read_csv('Python Course/Data for Python Course/salaries.csv') #importing salaries dataset
sales = pd.read_csv('Python Course/Data for Python Course/web_sales.csv') #importing web_sales dataset
eth = pd.read_csv('Python Course/Data for Python Course/Ethereum Historical Data.csv', index_col=0)         # If we already know what the index in the dataset is going to be then can be add here. 
```

# Creating Bar Plots and Adding Annotations

# Structuring Visualizations with Figures

```python
# using etherium dataset
plt.figure(figsize = (4,3), dpi = 120)        # dpi = dot per inch
plt.plot(eth.index, eth['Price'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Ethereum Price Chart')
plt.xticks(rotation = 45)
plt.show()

plt.figure(figsize = (4,3), dpi = 120)        # dpi = dot per inch
plt.plot(eth.index, eth['High'], color = 'orange')
plt.xlabel('Date')
plt.ylabel('High')
plt.title('Ethereum Price Chart')
plt.xticks(rotation = 45)
plt.show()

plt.figure(figsize = (4,3), dpi = 120)        # dpi = dot per inch
plt.plot(eth.index, eth['Low'], color = 'red')
plt.xlabel('Date')
plt.ylabel('Low')
plt.title('Ethereum Price Chart')
plt.xticks(rotation = 45)
plt.show()
```

# Organize Plots using Subplot

```python
# using etherium dataset
plt.figure(figsize = (10,3))   

plt.subplot(1,3,1)
plt.plot(eth.index, eth['Price'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Ethereum Closing Price Chart')
plt.xticks(rotation = 45)

plt.subplot(1,3,2)
plt.plot(eth.index, eth['High'], color = 'orange')
plt.xlabel('Date')
plt.ylabel('High')
plt.title('Ethereum High Price Chart')
plt.xticks(rotation = 45)

plt.subplot(1,3,3)
plt.plot(eth.index, eth['Low'], color = 'red')
plt.xlabel('Date')
plt.ylabel('Low')
plt.title('Ethereum Low Price Chart')
plt.xticks(rotation = 45)
plt.tight_layout()  # helps in adjusting labels
plt.show()
```

# Visualizing Sales Data - Part One

```python
bike_sales = sales[sales['product_subcategory'].isin(['Road Bikes', 'Mountain Bikes', 'Touring Bikes'])]
```

```python
bike_sales_grp = bike_sales.groupby('product_subcategory')['SalesAmount'].sum().reset_index()
```

```python
bike_sales_grp
```

```python
accessory_sales = sales[~sales['product_subcategory'].isin(['Road Bikes', 'Mountain Bikes', 'Touring Bikes'])]
```

```python
accessory_sales_grp = accessory_sales.groupby('product_subcategory')['SalesAmount'].sum().reset_index().sort_values(by = 'SalesAmount')
```

```python
accessory_sales_grp
```

```python
plt.pie(bike_sales_grp['SalesAmount'], labels = bike_sales_grp['product_subcategory'])
```

```python
plt.barh(accessory_sales_grp['product_subcategory'], accessory_sales_grp['SalesAmount'])
```

# Visualizing Sales Data - Part Two

```python
sales.info()
```

```python
sales['OrderDate'] = pd.to_datetime(sales['OrderDateKey'], format = '%Y%m%d')
sales['MonthYear'] = pd.to_datetime(sales['OrderDate'].dt.strftime('%Y-%m'))
```

```python
sales['OrderDate'].info()
```

```python
monthly_sales = sales.groupby('MonthYear')['SalesAmount'].sum().reset_index()
daily_sales = sales.groupby('OrderDate')['SalesAmount'].sum().reset_index()
```

```python
plt.plot(monthly_sales['MonthYear'], monthly_sales['SalesAmount'])
plt.show()
```

```python
plt.plot(daily_sales['OrderDate'], daily_sales['SalesAmount'])
plt.show()
```

```python
country_sales = sales.groupby('SalesTerritoryCountry')['SalesAmount']\
                .sum().reset_index().sort_values(by='SalesAmount',ascending=False)
```

```python
plt.bar(country_sales['SalesTerritoryCountry'], country_sales['SalesAmount'])
```

# Formatting Sales Plots - Part One

```python
plt.style.available
plt.style.use('dark_background')
plt.rcdefaults()
```

```python
plt.figure(figsize = (12,8))

plt.subplot(2,2,1)
#Line chart
plt.plot(monthly_sales['MonthYear'], monthly_sales['SalesAmount'], marker = '.', linestyle = '--', linewidth = 2.5, color = 'orange')
plt.xticks(rotation=45, fontsize = 10, color = 'black')
plt.title('Sales by Month', fontweight = 'bold')
plt.xlabel('Month-Year', fontsize = 10)
plt.ylabel('Sales', fontsize = 10)

plt.subplot(2,2,2)
#Horizontal Bar chart
plt.barh(accessory_sales_grp['product_subcategory'], accessory_sales_grp['SalesAmount'], color = 'green')
plt.title('Accessory Sales', fontweight = 'bold')

plt.subplot(2,2,3)
#Pie Chart
plt.pie(bike_sales_grp['SalesAmount'], labels = bike_sales_grp['product_subcategory'], autopct = '%1.1f%%', colors = ['indianred', 'teal', 'lightskyblue'])
plt.title('Bike Sales', fontweight = 'bold')


plt.subplot(2,2,4)
#Bar Chart
plt.bar(country_sales['SalesTerritoryCountry'], country_sales['SalesAmount'], color = 'steelblue')
plt.xticks(rotation = 45)
plt.title('Country Sales', fontweight = 'bold')
plt.xlabel('Country')
plt.ylabel('Sales')
plt.tight_layout()
```
